```python
# First deal with the AGe null values
train_data['Age'].fillna(train_data['Age'].median(),inplace=True)
test_data['Age'].fillna(test_data['Age'].median(),inplace=True)
```

```python
train_data.drop(['Cabin', 'Ticket', 'Name', 'PassengerId'], axis=1, inplace=True)
test_data.drop(['Cabin', 'Ticket', 'Name', 'PassengerId'], axis=1, inplace=True)
```

```python
# fare has null values
train_data['Fare'].fillna(train_data['Fare'].median(),inplace=True)
```

```python
train_data.isnull().sum()
```

```python
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)
```

```python
train_data['Embarked'].fillna(train_data['Embarked'].mode(), inplace=True)
test_data['Embarked'].fillna(test_data['Embarked'].mode(), inplace=True)
```

```python
test_data.isnull().sum()
```

```python
train_data=pd.get_dummies(train_data,columns=['Embarked'],drop_first=True)
test_data=pd.get_dummies(test_data,columns=['Embarked'],drop_first=True)
```

```python
train_data
```

```python
# now convert sex male:0 and female:1
train_data['Sex']=train_data['Sex'].map({'male':0,'female':1})
test_data['Sex']=test_data['Sex'].map({'male':0,'female':1})
```

```python
test_data['Sex']
```

```python
X=train_data.drop('Survived',axis=1)
```

```python
y=train_data['Survived']
```

```python
y
```

```python
y.ndim
```

```python
from sklearn.model_selection import train_test_split

# Split into training and validation sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression


# Train the model

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=model.predict(X_test)
accuracy=accuracy_score(y_test, y_pred)
print("Accuracy:",accuracy)
```
