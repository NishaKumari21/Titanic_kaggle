```python
import pandas as pd

train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

```

```python
# Fill missing values in 'Age' with the median age
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)

# Fill missing values in 'Embarked' with the most common value
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
test_data['Embarked'].fillna(test_data['Embarked'].mode()[0], inplace=True)

# Fill missing values in 'Fare' with the median fare
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)

# Drop unnecessary columns
train_data.drop(['Cabin', 'Ticket', 'Name', 'PassengerId'], axis=1, inplace=True)
test_data.drop(['Cabin', 'Ticket', 'Name', 'PassengerId'], axis=1, inplace=True)


```

```python
# Convert 'Sex' column to binary (0 = male, 1 = female)
train_data['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
test_data['Sex'] = test_data['Sex'].map({'male': 0, 'female': 1})

# Convert 'Embarked' column into dummy variables
train_data = pd.get_dummies(train_data, columns=['Embarked'], drop_first=True)
test_data = pd.get_dummies(test_data, columns=['Embarked'], drop_first=True)
```

```python
train_data.head()
```

```python
# Split features and target variable
X = train_data.drop('Survived', axis=1)
y = train_data['Survived']
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Split into training and validation sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Train the model

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
# Validate the model
from sklearn.metrics import accuracy_score
y_pred1 = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred1)

print(f'Validation Accuracy: {accuracy:.4f}')
```

```python
# Make predictions using logistic
test_prediction1 = model.predict(test_data)

# Create a submission file
submission = pd.DataFrame({'PassengerId': range(892, 1310), 'Survived': test_prediction1})
submission.to_csv('logistic_titanic_submission.csv', index=False)

print("Submission file 'logistic_titanic_submission.csv' is ready!")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
# Split into training and validation sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Train the model
model = RandomForestClassifier(n_estimators=200, random_state=42)
model.fit(X_train, y_train)
```

```python
# Validate the model
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f'Validation Accuracy: {accuracy:.8f}')
```

```python
# Make predictions
test_predictions = model.predict(test_data)

# Create a submission file
submission = pd.DataFrame({'PassengerId': range(892, 1310), 'Survived': test_predictions})
submission.to_csv('random_titanic_submission.csv', index=False)

print("Submission file 'Random_titanic_submission.csv' is ready!")
```
